### Precipitation Data for Beja

```python
df_prec_beja_raw = pd.read_csv('/content/prec-Beja-raw.csv')
display(df_prec_beja_raw.info())
```

```python
# Filter df_prec_faro (Precipitation Faro)
df_prec_faro_filtered = df_prec_faro[df_prec_faro['year'] >= 2000]
print("\n--- Filtered Precipitation Faro (df_prec_faro_filtered) ---")
display(df_prec_faro_filtered.head())
display(df_prec_faro_filtered.info())

# Filter df_temp_beja_raw (Temperature Beja)
# Reload df_temp_beja_raw to ensure it's defined
df_temp_beja_raw = pd.read_csv('/content/temp-Beja-raw.csv')
df_temp_beja_raw['date'] = pd.to_datetime(df_temp_beja_raw['date'], format='%m/%Y')
df_temp_beja_raw_filtered = df_temp_beja_raw[df_temp_beja_raw['date'].dt.year >= 2000]
print("\n--- Filtered Temperature Beja (df_temp_beja_raw_filtered) ---")
display(df_temp_beja_raw_filtered.head())
display(df_temp_beja_raw_filtered.info())

# Filter df_temp_faro (Temperature Faro)
df_temp_faro['date'] = pd.to_datetime(df_temp_faro['date'], format='%m/%Y')
df_temp_faro_filtered = df_temp_faro[df_temp_faro['date'].dt.year >= 2000]
print("\n--- Filtered Temperature Faro (df_temp_faro_filtered) ---")
display(df_temp_faro_filtered.head())
display(df_temp_faro_filtered.info())

# Filter df_prec_beja_raw (Precipitation Beja)
df_prec_beja_raw_filtered = df_prec_beja_raw[df_prec_beja_raw['year'] >= 2000]
print("\n--- Filtered Precipitation Beja (df_prec_beja_raw_filtered) ---")
display(df_prec_beja_raw_filtered.head())
display(df_prec_beja_raw_filtered.info())
```

 ## Juntar dados da percipitação

Alentejo (Beja)

```python
prec_beja = (
    df_prec_beja_raw_filtered[['year', 'Annual']]
    .copy()
)

prec_beja['region'] = 'Alentejo'

prec_beja.rename(
    columns={'Annual':'precipitation_mm'},
    inplace=True
)
```

Algarve (Faro)


```python
prec_faro = (
    df_prec_faro_filtered[['year', 'Annual']]
    .copy()
)

prec_faro['region'] = 'Algarve'

prec_faro.rename(
    columns={'Annual':'precipitation_mm'},
    inplace=True
)
```

Juntar precipitação

```python
precipitation = pd.concat(
    [prec_beja, prec_faro],
    ignore_index=True
)

display(precipitation.head(40))
```

## Juntar dados da temperatura

Criar coluna ano:

```python
df_temp_beja_raw_filtered['year'] = (
    df_temp_beja_raw_filtered['date']
    .dt.year
)
```

Agregar os dados

```python
temp_beja = (
    df_temp_beja_raw_filtered
    .groupby('year')
    .agg(
        mean_temp=('tmed', 'mean'),
        mean_tmax=('tmax', 'mean'),
        mean_tmin=('tmin', 'mean')
    )
    .reset_index()
)
```

Adicionar região:

```python
temp_beja['region'] = 'Alentejo'
```

Agora o mesmo para Faro

```python
df_temp_faro_filtered['year'] = (
    df_temp_faro_filtered['date']
    .dt.year
)

temp_faro = (
    df_temp_faro_filtered
    .groupby('year')
    .agg(
        mean_temp=('tmed', 'mean'),
        mean_tmax=('tmax', 'mean'),
        mean_tmin=('tmin', 'mean')
    )
    .reset_index()
)

temp_faro['region'] = 'Algarve'
```

Juntar a temperatura das duas regioes

```python
temperature = pd.concat(
    [temp_beja, temp_faro],
    ignore_index=True
)

display(temperature.head(40))
```

 ## Construir a tabela climática final

Agora fazemos o merge:

```python
climate = pd.merge(
    precipitation,
    temperature,
    on=['year', 'region'],
    how='outer' # Changed from 'inner' to 'outer'
)

display(climate.head(40))
```

## Guardar

```python
climate.to_csv(
    'climate_final.csv',
    index=False
)
```

```python
irr_total = irr[
    irr["land_use_type"] == "Total"
].copy()
```

```python
irr_total = irr_total[
    ["year", "region", "irrigable_area_ha"]
]
```

```python
irr_total.head()
```

```python
irr_total.sort_values(["region", "year"])
```

```python
irr_total.groupby(["year", "region"]).size()
```

## Garantir que os tipos estão corretos

```python
irr_total["year"] = irr_total["year"].astype(int)

irr_total["irrigable_area_ha"] = pd.to_numeric(
    irr_total["irrigable_area_ha"],
    errors="coerce"
)
```

Criar uma linha para todos os anos (2000-2018)

```python
years = range(2000, 2019)

irr_interp = []

for region in irr_total["region"].unique():

    temp = (
        irr_total[irr_total["region"] == region]
        .set_index("year")
        .reindex(years)
    )

    temp["region"] = region

    temp["irrigable_area_ha"] = (
        temp["irrigable_area_ha"]
        .interpolate(method="linear")
    )

    temp = (
        temp
        .reset_index()
        .rename(columns={"index": "year"})
    )

    irr_interp.append(temp)

irr_interp = pd.concat(
    irr_interp,
    ignore_index=True
)
```

```python
irr_interp.head(25)
```

## Filtrar ambas as tabelas para 2005-2018

```python
prod = prod[
    (prod["year"] >= 2005) &
    (prod["year"] <= 2018)
].copy()

irr_interp = irr_interp[
    (irr_interp["year"] >= 2005) &
    (irr_interp["year"] <= 2018)
].copy()
```

## Merge de prod com irr_interp

```python
agri = pd.merge(
    prod,
    irr_interp,
    on=["year", "region"],
    how="left"
)
```

```python
agri.tail(100)
```

```python
agri = agri[agri["species"] != "Uva para vinho (Vinhos sem certificação)"]
```

```python
agri.head(100)
```

```python
agri.info()
```

## se não há área → provavelmente não há cultivo

```python
agri["area_ha"] = agri["area_ha"].fillna(0)
```

## Produtividade só faz sentido ser 0 se área = 0

```python
agri.loc[agri["area_ha"] == 0, "production_t"] = 0
```

## recalcular productivity_kg_ha

```python
agri["productivity_kg_ha"] = (
    agri["production_t"] * 1000 / agri["area_ha"]
)
```

```python
agri.loc[agri["area_ha"] == 0, "productivity_kg_ha"] = 0
```
